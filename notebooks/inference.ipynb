```python
import os

def process_large_files(directory_path):
    print(f"{'File Name':<20} | {'Word Count':<12} | {'Preview (First 25 Words)'}")
    print("-" * 100)
    
    total_words_global = 0

    # Sorting ensures they appear in order (ne_dedup_1, 2, 3...)
    filenames = sorted(os.listdir(directory_path))

    for filename in filenames:
        if filename.endswith(".txt"):
            file_path = os.path.join(directory_path, filename)
            word_count = 0
            preview_words = []
            
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    for line in f:
                        words = line.split()
                        # If we haven't collected 25 words yet, add them to our preview list
                        if len(preview_words) < 25:
                            preview_words.extend(words)
                        
                        word_count += len(words)
                
                total_words_global += word_count
                # Clean up the preview to exactly 25 words and join them
                preview_str = " ".join(preview_words[:25])
                print(f"{filename:<20} | {word_count:<12,} | {preview_str}...")
                
            except Exception as e:
                print(f"{filename:<20} | Error        | Could not process: {e}")

    print("-" * 100)
    print(f"{'GRAND TOTAL':<20} | {total_words_global:<12,}")

# Run it
process_large_files('./data/text/')
```

```python
import re
import unicodedata
from pathlib import Path

JUNK_CHARS = {
    '\u200d',  # Zero-width joiner
    '\u200c',  # Zero-width non-joiner
    '\u200b',  # Zero-width space
    '\ufeff',  # BOM
    '\u200e',  # Left-to-right mark
    '\u200f',  # Right-to-left mark
    '\xa0',    # Non-breaking space
}

# Punctuation to strip (not in grapheme vocab)
STRIP_CHARS = re.compile(r'[।॥,\.!\?\"\'\(\)\[\]\{\}:;/\\\-–—|@#\$%\^&\*\+=<>`~]+')

# Any remaining ASCII except digits
ASCII_NON_DIGIT = re.compile(r'[a-zA-Z]')

def clean_line(text: str) -> str:
    # Remove junk unicode
    text = ''.join(c for c in text if c not in JUNK_CHARS)
    # NFC normalize (important for Devanagari)
    text = unicodedata.normalize('NFC', text)
    # Strip punctuation
    text = STRIP_CHARS.sub(' ', text)
    # Remove stray Roman letters
    text = ASCII_NON_DIGIT.sub(' ', text)
    # Collapse whitespace
    text = ' '.join(text.split())
    return text.strip()

input_files = sorted(Path('data/text/').glob('ne_dedup_*.txt'))
output_file = Path('nepali_corpus_clean.txt')

total_lines = 0
total_words = 0

with open(output_file, 'w', encoding='utf-8') as out:
    for fpath in input_files:
        print(f"Processing {fpath.name}...")
        with open(fpath, encoding='utf-8') as f:
            for line in f:
                cleaned = clean_line(line)
                if len(cleaned) < 5:  # skip near-empty lines
                    continue
                out.write(cleaned + '\n')
                total_lines += 1
                total_words += len(cleaned.split())

print(f"Done: {total_lines:,} lines | {total_words:,} words")
print(f"Output: {output_file}")
```

```python
from pathlib import Path
# make sure your tokenizer is importable
import sys
sys.path.insert(0, str(Path('.').resolve()))
from src.tokenizer import *

tokenizer = GraphemeTokenizer.load('./data/tokenizer.json')

input_file  = Path('./data/nepali_corpus_clean.txt')
output_file = Path('./data/nepali_corpus_grapheme.txt')

def line_to_graphemes(line: str, tokenizer) -> str:
    line = line.strip()
    if not line:
        return ''
    ids = tokenizer.encode(line)  # handles cleaning, grapheme splitting, space insertion
    gs = [tokenizer.id2char[i] for i in ids]
    # filter out unk tokens (corpus words not in ASR vocab)
    gs = [g for g in gs if g != tokenizer.UNK_TOKEN]
    return ' '.join(gs)

test_lines = [
    "शाखाहरूमा विभक्त रहन्छ",
    "दक्षिण कोरियाले सेनालाई",
    "महिलाहरू व्यवसायी बन्ने",
]
for line in test_lines:
    print(repr(line_to_graphemes(line, tokenizer)))

total = 0
with open(input_file, encoding='utf-8') as fin, \
     open(output_file, 'w', encoding='utf-8') as fout:
    for line in fin:
        converted = line_to_graphemes(line, tokenizer)
        if converted:
            fout.write(converted + '\n')
            total += 1
        if total % 100_000 == 0:
            print(f"  {total:,} lines done")

print(f"Done: {total:,} lines → {output_file}")
```
